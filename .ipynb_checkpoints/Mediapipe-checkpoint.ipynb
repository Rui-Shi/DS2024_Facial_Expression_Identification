```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import mediapipe as mp
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf

# Initialize MediaPipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=1, refine_landmarks=True)

train_dir = '/Users/sherrywang/Desktop/fer2013/train'

# Define an ImageDataGenerator without augmentation for loading
datagen = ImageDataGenerator(rescale=1./255)  # Rescale pixel values to [0, 1], normalization

# Load the images and labels from the directory
training = datagen.flow_from_directory(train_dir, target_size=(48, 48), color_mode='grayscale', batch_size=16, class_mode='categorical', shuffle=False)

print(training.class_indices)

def extract_mediapipe_features(image):
    """
    Extract facial landmarks using MediaPipe Face Mesh and return the feature vector.
    Parameters:
    - image: The input image (RGB).
    
    Returns:
    - Flattened 3D landmarks (x, y, z) as feature array.
    """
    # Ensure image is of type uint8 (0-255 range)
    image_uint8 = np.array(image, dtype=np.uint8)
    
    # Process the image with MediaPipe Face Mesh
    results = face_mesh.process(image_uint8)
    

    if not results.multi_face_landmarks:
        return None

    else: # Loop through detected faces
        for face_landmarks in results.multi_face_landmarks:
                    # Draw the face landmarks on the original image
                    mp.solutions.drawing_utils.draw_landmarks(
                        image=  image_uint8, 
                        landmark_list=face_landmarks,
                        connections=mp_face_mesh.FACEMESH_TESSELATION,  # Facial mesh connections
                        landmark_drawing_spec=None,  # Use default landmark drawing spec
                        connection_drawing_spec=mp.solutions.drawing_styles.get_default_face_mesh_tesselation_style()
                    )

                # Display the image with landmarks (optional)
                # Comment this out if you do not want to display each image
        cv2.imshow('Face Mesh', image_uint8)
             

    # Extract landmark coordinates for the first detected face
    landmarks = results.multi_face_landmarks[0]
   #print(results.multi_face_landmarks)
    #print('over')
    #print(landmarks)
   # print('over1')

    feature_vector = []
    
    for landmark in landmarks.landmark:
        x = landmark.x  # x coordinate
        y = landmark.y  # y coordinate
        z = landmark.z  # z coordinate
        feature_vector.append([x, y, z])

    return np.array(feature_vector).flatten()  # Flatten into a 1D feature vector

if __name__ == '__main__':   
    X_mediapipe_features = []

    for i in range(len(training)):
        X_batch, y_batch = next(training)

        # Convert each grayscale image in the batch to RGB
        for img in X_batch:
            img_rgb = tf.image.grayscale_to_rgb(tf.convert_to_tensor(img))  # Convert grayscale to RGB
            img_rgb = np.uint8(img_rgb.numpy() * 255.0)  # Convert to uint8 in [0, 255] range

            # Extract MediaPipe features
            mediapipe_features = extract_mediapipe_features(img_rgb)
            if mediapipe_features is not None:
                X_mediapipe_features.append(mediapipe_features)
                print("it works")
            else:
                print("No face detected in the image.")

```

```python

# Example of plotting an image with its landmarks
img_example, label_example = next(training)
img_rgb_example = tf.image.grayscale_to_rgb(tf.convert_to_tensor(img_example[0]))  # Convert grayscale to RGB
img_rgb_example = np.uint8(img_rgb_example.numpy() * 255.0)  # Convert to uint8 in [0, 255] range

mediapipe_example = extract_mediapipe_features(img_rgb_example)

    # Plot original image and landmark overlay
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_example[0], cmap='gray')  # Display the original grayscale image
plt.title('Original Image')

    # Check if landmarks were detected
if mediapipe_example is not None:
    plt.subplot(1, 2, 2)
     # Flip both x and y coordinates for 180-degree clockwise rotation
    x_coords_rotated = [x for x in mediapipe_example[::3]]  # Flip x coordinates
    y_coords_rotated = [-y for y in mediapipe_example[1::3]]  # Flip y coordinates
    
    # Plot the rotated landmarks
    plt.scatter(x_coords_rotated, y_coords_rotated, c='red', s=10)  # Plot x and y coordinates
    plt.title('MediaPipe Landmarks')

    plt.show()
```
